```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('knnDataSet.csv', index_col=0)
```

```python
plt.scatter(df["x"], df["y"], color='b')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

This scatter plot suggests that data has 2 or 4 groups. Also, it seems that x and y are positively correlated in some of this groups.

```python
plt.scatter(df["x"], df["y"], c=df["L"])
plt.scatter(df.groupby('L')['x'].mean(), df.groupby('L')['y'].mean(), color='r')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
train, test = train_test_split(df, test_size=0.25)
```

```python
knn = KNeighborsClassifier(n_neighbors=3).fit(train[["x","y"]], train["L"])

print(confusion_matrix(test["L"], knn.predict(test[["x","y"]])))
print("\n", classification_report(test["L"], knn.predict(test[["x","y"]])))
print("\nAccuracy: ", accuracy_score(test["L"], knn.predict(test[["x","y"]])))
```

```python
knn = KNeighborsClassifier(n_neighbors=1).fit(train[["x","y"]], train["L"])

print(confusion_matrix(test["L"], knn.predict(test[["x","y"]])))
print("\n", classification_report(test["L"], knn.predict(test[["x","y"]])))
print("\nAccuracy: ", accuracy_score(test["L"], knn.predict(test[["x","y"]])))
```
